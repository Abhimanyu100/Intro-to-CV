```python
"""
It is recommended to tweak around images using NumPy arrays, rather than processing them pixel-by-pixel
which would delay any operation done on the images.
The OpenCV module is imported as cv2
For display of images, matplotlib is a standard library

"""

%matplotlib inline
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
"""

Images are read using "imread"

"""

img = cv2.imread("baboon256.tif",0)
```

```python
"""
    Displaying the Image 
"""
plt.imshow(img)
plt.show()
```

cv2 has its GUI feature as well, which is used for displaying images.
It can be done using -
cv2.imshow("Name on the header",img)
However, there are a lot of bugs in that. For example, once the image window opens, it never closes. In Linux or OSX systems, often the python system hangs.

Another method that is used is cv2.waitKey()
It is used to generate interrupts for the GUI Window

Example :

cv2.imshow('image', img)

key = cv2.waitKey(0)

if k==ord('s'):

    cv2.imwrite('messigray.png',img)
    
    cv2.destroyAllWindows()
    

''' Note - If you are using a 64 bit machine

    modify the command as :
    
    k = cv2.waitKey(0) & 0xFF '''

```python
"""
To check the pixel values of this GrayScale image. A GrayScale image would have pixel values in a range of 0-255.
"""

print img.max()
print img.min()
```

```python
"""
img.shape will return a tuple which shows the rows, columns and channels (if image is color).
img[x,y] will return intensity value if its a GrayScale image.
img.size displays the total number of pixels!
img.dtype displays the datatype of the image. Very important!

"""

print img.shape

print img[78,78] 

print img.size

print img.dtype
```
